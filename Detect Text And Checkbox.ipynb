```python
import os
```

```python
import pandas as pd

df = pd.read_csv("processed_text_data.csv", index_col='index')

df.head()
```

```python
import pickle
import numpy as np
from PIL import Image
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

# Load and preprocess images
def load_image(image_path):
    img = Image.open(image_path).convert('L')  # Convert to grayscale
    img = img.resize((25, 25))  # Resize image
    img_array = np.array(img).flatten()  # Flatten image
    return img_array

with open('checkbox_model.pkl', 'rb') as f:
    classifier = pickle.load(f)
```

```python
column_name = "race_of_household"
base_path = "data/race_of_household-1128-cut"
checkbox_subfolders = ["white", "negro", "mexican", "oriental", "other"]

df[column_name] = None


all_file_names = os.listdir(os.path.join(base_path, checkbox_subfolders[0]))
all_file_names = [f for f in all_file_names if f.endswith(('.png', '.jpg', '.jpeg', '.tiff', '.bmp', '.gif'))]

for index in all_file_names:
    for folder in checkbox_subfolders:
        image_path = os.path.join(base_path, folder, index)
        if not os.path.exists(image_path):
            print(f"{image_path} does not exist")
            prediction = [0]
        else:
            prediction = classifier.predict([load_image(image_path)])
        if prediction[0] == 1:
            print(folder)
            race = folder
            break
    df.at[index, column_name] = race

df.to_csv("processed_text_data_001.csv")
df.head(10)
    
```
